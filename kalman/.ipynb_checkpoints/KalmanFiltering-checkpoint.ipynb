Kalman filtering
======================================================================

This is code implements the example given in pages 11-15 of [An
Introduction to the Kalman
Filter](https://www.cs.unc.edu/~welch/media/pdf/kalman_intro.pdf) by Greg
Welch and Gary Bishop, University of North Carolina at Chapel Hill,
Department of Computer Science.

Download from https://scipy-cookbook.readthedocs.io/items/KalmanFiltering.html in 2019-10-26.

Comments and explanation added, and reformatted by chenxy.

```python
# Kalman filter example demo in Python

# A Python implementation of the example given in pages 11-15 of "An
# Introduction to the Kalman Filter" by Greg Welch and Gary Bishop,
# University of North Carolina at Chapel Hill, Department of Computer
# Science, TR 95-041,
# https://www.cs.unc.edu/~welch/media/pdf/kalman_intro.pdf

# by Andrew D. Straw

import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10, 8)
```

```python
# intial parameters
n_iter = 50
sz = (n_iter,) # size of array. This example handles scaler parameter estimation, hence 1-D tensor tensor is used throughout.
x = -0.37727 # truth value (typo in example at top of p. 13 calls this z)
z = np.random.normal(x,0.1,size=sz) # observations (normal about x, sigma=0.1)

A = 1    # Added by chenxy for clarity.
B = 0    # Added by chenxy for clarity.
H = 1    # Added by chenxy for clarity. Not appearing in the following calculation, because it is just ignored.
Q = 1e-5 # process variance. Find the explanation why not just set it to zero, in the original paper, page12
R        = 0.1**2 # estimate of measurement variance, change to see effect

# allocate space for arrays
u    =np.zeros(sz)      # driving force. zero in this example.
xhat =np.zeros(sz)      # a posteri estimate of x
P    =np.zeros(sz)      # a posteri error estimate
xhatminus=np.zeros(sz)  # a priori estimate of x
Pminus   =np.zeros(sz)  # a priori error estimate
K        =np.zeros(sz)  # gain or blending factor

```

```python
# intial guesses
xhat[0] = 0.0
P[0] = 1.0

# Recursive estimation
for k in range(1,n_iter):
    # time update
    xhatminus[k] = A * xhat[k-1] + B * u[k-1]
    Pminus[k] = P[k-1]+Q
    
    # measurement update
    K[k]    = Pminus[k]/( Pminus[k]+R )
    xhat[k] = xhatminus[k]+K[k]*(z[k]-xhatminus[k])
    P[k]    = (1-K[k])*Pminus[k]
```

```python
# Naive estimation based on sample mean average 
x_sma = np.zeros(sz)      # sample-mean-average estimation of x
x_sma[0] = z[0]
for k in range(1,len(x_sma)):
    x_sma[k] = np.mean(z[:k+1])    
```

```python
# Plot for visualization of 
plt.figure()
plt.plot(z,'k+',label='noisy measurements')
plt.plot(xhat,'b-',label='a posteri estimate')
plt.plot(x_sma,'go',label='sample mean average estimate')
plt.axhline(x,color='g',label='truth value')
plt.legend()
plt.title('Estimate vs. iteration step', fontweight='bold')
plt.xlabel('Iteration')
plt.ylabel('Voltage')

plt.figure()
plt.plot(xhat-x_sma,label = 'difference between two estimations')
plt.title('difference between kalman estimation and naive sample-mean-average measurement', fontweight='bold')
plt.axhline(0,color='g',label='zero')
plt.legend()
plt.xlabel('Iteration')
plt.ylabel('Voltage')
```

Observation: 

In this simple static case, sample-mean-average estimation and kalman filtering estimation are very close, except the intialization stage. 

```python
plt.figure()
valid_iter = range(1,n_iter) # Pminus not valid at step 0
plt.plot(valid_iter,Pminus[valid_iter],label='a priori error estimate')
plt.plot(valid_iter,P[valid_iter],label='a posterori error estimate')
#plt.title('Estimated $\it{\mathbf{a \ priori}}$ error vs. iteration step', fontweight='bold')
plt.title('Estimated error vs. iteration step', fontweight='bold')
plt.legend()
plt.xlabel('Iteration')
plt.ylabel('$(Voltage)^2$')
plt.setp(plt.gca(),'ylim',[0,.01])
plt.show()
```

## Effect of measurement noise

If we decrese the measurement noise setting, then the mearsurement value will get larger weight, so the estimation will try to tracking the measurement variation quicklier, and hence larger variance, as shown below.

```python
R = 0.01 ** 2  # Smaller a-priori measurement noise

# intial guesses
xhat[0] = 0.0
P[0] = 1.0

# Recursive estimation
for k in range(1,n_iter):
    # time update
    xhatminus[k] = A * xhat[k-1] + B * u[k-1]
    Pminus[k] = P[k-1]+Q
    
    # measurement update
    K[k]    = Pminus[k]/( Pminus[k]+R )
    xhat[k] = xhatminus[k]+K[k]*(z[k]-xhatminus[k])
    P[k]    = (1-K[k])*Pminus[k]
    
plt.figure()
plt.plot(z,'k+',label='noisy measurements')
plt.plot(xhat,'b-',label='a posteri estimate')
plt.axhline(x,color='g',label='truth value')
plt.legend()
plt.title('Estimate vs. iteration step', fontweight='bold')
plt.xlabel('Iteration')
plt.ylabel('Voltage')
```

On the other hand, if we increse the measurement noise setting, then the system will give smaller weight to mearsurement,so the estimation will try to tracking the measurement variation slowlier, and hence will need longer time to converge. For this example, because the parameter-to-be-estimated is static, given enough iteration, the system will converge to the ground-truth finally, with smaller variance, but if the parameter-to-be-estimated is dynamically change over time, the system may be unable to track the change.

```python
# intial parameters
n_iter = 200
sz = (n_iter,) # size of array. This example handles scaler parameter estimation, hence 1-D tensor tensor is used throughout.
x = -0.37727 # truth value (typo in example at top of p. 13 calls this z)
z = np.random.normal(x,0.1,size=sz) # observations (normal about x, sigma=0.1)

A = 1    # Added by chenxy for clarity.
B = 0    # Added by chenxy for clarity.
Q = 1e-5 # process variance. Find the explanation why not just set it to zero, in the original paper, page12

# allocate space for arrays
u    =np.zeros(sz)      # driving force. zero in this example.
xhat =np.zeros(sz)      # a posteri estimate of x
P    =np.zeros(sz)      # a posteri error estimate
xhatminus=np.zeros(sz)  # a priori estimate of x
Pminus   =np.zeros(sz)  # a priori error estimate
K        =np.zeros(sz)  # gain or blending factor
R = 1 ** 2  # Smaller a-priori measurement noise

# intial guesses
xhat[0] = 0.0
P[0] = 1.0

# Recursive estimation
for k in range(1,n_iter):
    # time update
    xhatminus[k] = A * xhat[k-1] + B * u[k-1]
    Pminus[k] = P[k-1]+Q
    
    # measurement update
    K[k]    = Pminus[k]/( Pminus[k]+R )
    xhat[k] = xhatminus[k]+K[k]*(z[k]-xhatminus[k])
    P[k]    = (1-K[k])*Pminus[k]
    
plt.figure()
plt.plot(z,'k+',label='noisy measurements')
plt.plot(xhat,'b-',label='a posteri estimate')
plt.axhline(x,color='g',label='truth value')
plt.legend()
plt.title('Estimate vs. iteration step', fontweight='bold')
plt.xlabel('Iteration')
plt.ylabel('Voltage')
```
